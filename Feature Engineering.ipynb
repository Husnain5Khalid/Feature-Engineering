## Feature Engineering

```python
df.head()
```

```python
df["experience_years"].unique()
```

```python
def experience_level(x):
    if x <= 2:
        return "Junior"
    elif x<= 5:
        return "Mid"
    else:
        return "Senior"
df["Experience_Level"] = df["experience_years"].apply(experience_level)
```

```python
df.head()
```

```python
df["education_level"].unique()
```

```python
def education(x):
    if x == "High School" or x == "Diploma":
        return "Basic education"
    else:
        return "Higher Education"
df["Education_status"] = df["education_level"].apply(education)
df.head()
```

```python
df["skills_count"].unique()
```

```python
def skills_level(x):
    if x <= 5:
        return "Low"
    elif x <= 10:
        return "Medium"
    else:
        return "High"
df["Skills_level"] = df["skills_count"].apply(skills_level)
df.head()
```

```python
df["Skills_per_year"] = df["skills_count"]/ (df["experience_years"] + 1)
df.head()
```

```python
df["certifications"].unique()
```

```python
df["location"].unique()
```

```python
df["salary based on skills&exp"]  = df["salary"] / (df["skills_count"] * (df["experience_years"] + 1))
df.head()
```
